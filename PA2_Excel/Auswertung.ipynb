```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = [6, 0, 1, 0, 0,
5, 1, 1, 0, 0,
1, 3, 2, 0, 1,
4, 2, 2, 0, 0,
2, 1, 2, 1, 1,
4, 0, 2, 0, 0,
1,1,1,1,3
]

df = pd.DataFrame(np.array(data).reshape(7,5), columns=['sehr kurz (<3min)',
'kurz(<5 min)',
'mittel(<8)',
'lang(<15min)',
'sehr lang (>15 min)'], index =['Grafik_1', 'Grafik_2', 'Grafik_3', 'Grafik_4', 'Grafik_5',
       'Grafik_6', 'Grafik_7'])

ax = df.plot(kind='bar', rot=0, xlabel='Grafik', ylabel='Anzahl an Einträgen', title='Vergleich der Länge der Grafikerstellung', figsize=(15, 10))

for c in ax.containers:

    # customize the label to account for cases when there might not be a bar section
    labels = [f'{h:0.1f}' if (h := v.get_height()) != 0 else '' for v in c ]

    # set the bar label
    ax.bar_label(c, labels=labels, fontsize=8, padding=3)

ax.legend(title='Durchführungsdauer', bbox_to_anchor=(1, 1.02), loc='upper left')
plt.show()
```

```python
liste_zeit=['27:12','12:23','77:42', '28:54','19:38','07:19','12:00','4:00','18:00','7:00','5:00','5:00', '4:45','4:15']
```

```python
def time_to_decimal(time):
  # Teile die Zeit in Minuten und Sekunden auf
  minutes, seconds = time.split(":")

  # Wandle Minuten und Sekunden in Dezimalzahlen um
  decimal_time = int(minutes) + int(seconds) / 60

  return round(decimal_time,2)
```

```python
f_list=[]
for el in liste_zeit:
    f_list.append(time_to_decimal(el))
```

```python
def convert_to_time(decimal_time):
    minutes = int(decimal_time)
    seconds = int((decimal_time - minutes) * 60)
    return f"{minutes}:{seconds:02d}"

# Beispielnutzung
decimal_time = 12.34
print(convert_to_time(decimal_time))  # 12:20
```

```python
def Average(lst):
    return sum(lst) / len(lst)
 
# Driver Code

average = Average(f_list)
import statistics

# Printing average of the list
print("Average of the Time =", convert_to_time(Average(f_list)))
print("STD of the Time =", convert_to_time(statistics.stdev(f_list)))
print("Minimum of the Time =", convert_to_time(min(f_list)))
print("Max of the Time =", convert_to_time(max(f_list)))
```
